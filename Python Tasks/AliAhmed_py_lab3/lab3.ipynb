```python
class Queue:
    size = 0
    consumarsArr = []
    def insert(self,value):
        if len(self.consumarsArr) == self.size:
            raise IndexError("Queue is full !!")
        else:
            self.consumarsArr.append(value)
            # print(value, "is added to queue.")
    def pop(self):
        if len(self.consumarsArr) == 0:
            raise Exception("Queue is empty !!")
        else:
            self.consumarsArr.pop(0)
    def is_empty(self):
        if len(self.consumarsArr) == 0:
            return True
        else:
            return False
    def displayQueue(self):
        for i in range(len(self.consumarsArr)):
            print(self.consumarsArr[i])
    def setSize(self):
        self.size = int(input("Enter szie of Queue : "))

try:
    q1 = Queue()
    q1.setSize();
    print(q1.is_empty());
    q1.insert(1);
    q1.insert(9);
    q1.insert(3);
    q1.insert(4);
    q1.insert(5);
    q1.insert(6);
    print(q1.is_empty());
    q1.displayQueue();
except IndexError as ofe:
    print(ofe)
except Exception as e:
    print(e)
```
